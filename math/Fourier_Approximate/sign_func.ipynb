```python
import numpy as np 
import pandas as pd  
import plotly.express as px  
```

```python
N=int(input())
sum1 = 0
x_list = np.arange(-np.pi + 0.02, np.pi-0.02, 0.01)
y_list = [np.sign(x) for x in x_list]
func_df = pd.DataFrame({'x' :x_list,
                        'y': y_list,
                        'Function': 'Original Function'})
df = pd.DataFrame({'x':x_list,
                   'y': 0,
                   'Function': 'Fourier Series'})

for i, x_temp in enumerate(x_list):
    for k in range(0,N+1):
        sum1+=np.sin((2*k+1)*x_temp)/(2*k+1)
    sum1=sum1*(4/np.pi)
    df.at[i,'y'] = sum1
    sum1=0
df = pd.concat([df, func_df])
df
px.line(df, x='x', y='y', title=f'Fourier approximation with N = {N}', color='Function')
```
