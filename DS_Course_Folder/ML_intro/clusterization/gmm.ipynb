```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
plt.style.use('seaborn')
```

```python
data = pd.read_csv("../data/housing.csv")
data.head()
```

```python
sns.scatterplot(data['year'], data['price'])
```

```python
k_means = KMeans(n_clusters=3, random_state=42)
k_means.fit(data)
k_preds = k_means.predict(data)
```

```python
gmm = GaussianMixture(n_components=3, random_state=42, init_params='k-means++')
gmm.fit(data)
preds = gmm.predict(data)
sns.scatterplot(data['year'], data['price'], c=preds, cmap='Accent')
```
