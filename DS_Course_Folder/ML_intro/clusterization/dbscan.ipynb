```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
plt.style.use('seaborn')
```

```python
df = pd.read_csv('../data/pokemon.csv')
```

```python
clust = DBSCAN(eps=3, min_samples=3).fit(df[['Attack', 'Defense']])
```

```python
sns.scatterplot(df['Attack'], df['Defense'], c=clust.labels_, cmap='Blues')
```

```python
house_data = pd.read_csv('../data/housing.csv')
house_data
```

```python
house_clust = DBSCAN(eps=1.15, min_samples=2)
house_clust.fit(house_data)
sns.scatterplot(house_data['year'], house_data['price'], c=house_clust.labels_)
```
