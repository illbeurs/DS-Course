```python
import pandas as pd
data = pd.read_csv('../data/wine.csv')
```

```python
wines = data.copy()
wines['taster_name'].nunique()
```

```python
wines['price'].max()
```

```python
wines.info()
```

```python
dups = wines[wines.duplicated()]
display(wines.shape[0])
wines = wines.drop_duplicates()
display(wines.shape[0])
dups.shape[0]
```

```python
display(wines.info())
wines.isna().describe()
```

```python
thresh = 0.6*wines.shape[0]
wines = wines.dropna(thresh=thresh, axis=1)
wines.info()
```

```python
from plotly import express as px
display(px.box(wines, 'price'))
fill_data = {
    'country' : 'US',
    'designation': 'unknown',
    'price': wines['price'].mean(),
    'province' : wines['province'].mode()[0],
    'region_1': 'unknown',
    'taster_name' : 'unknown',
    'taster_twitter_handle' : 'unknown',
    'variety' : wines['variety'].mode()[0]
}
wines = wines.fillna(fill_data)

```

```python
wines.info()
```

```python
wines.to_csv('../data/wine_cleared.csv', index=False)
```

```python
from pandas_profiling import ProfileReport
profile = ProfileReport(wines, title='Wine Pandas Profiling Report')
profile
```
